```python

import pymupdf
from pathlib import Path
file_path = Path("/home/almino/Zotero/storage/LT355VSL/Joao-Cassino-Tese-de-Doutorado_v-deposito-1.pdf")

src = pymupdf.open(file_path)  # input PDF
```

```python
import pandas as pd

page = 25
page19 = list(src.pages(page - 1, page))[0]
# pd.DataFrame(page19.get_text('blocks'), columns=['x0', 'y0', 'x1', 'y1', 'lines', 'block_no', 'block_type'])
page_dict = page19.get_text('dict')
page_blocks = page_dict["blocks"]

pd.json_normalize(
    data=page_blocks,
)
```

```python

pd.json_normalize(
    data=page_blocks,
    record_path=['lines'],
    meta=['type']
)
```

```python

pd.json_normalize(
    data=page19.get_text('dict')["blocks"],
    record_path=['lines', 'spans'],
    meta=['type']
)
```
